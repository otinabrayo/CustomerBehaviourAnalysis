### My Cleaning proccess


1. Filled missing 'Ratings' values with the average rating, rounded to 1 decimal place
2. Filled missing categorical values with the most common value (mode)
3. Dropped unnecessary columns that are mostly empty or not useful for analysis
4. Dropped rows with null and cant be replaced because they are key to the Analysis
5. Saved the cleaned data

```python
CustomerData['Ratings'] = CustomerData['Ratings'].fillna(round(CustomerData['Ratings'].mean(), 1))
```

```python
categorical_cols = ['Gender', 'Shipping_Method', 'Payment_Method', 'Feedback',
                    'Order_Status', 'Product_Category', 'Product_Brand', 'Product_Type', 'Age']
for col in categorical_cols:
    if not CustomerData[col].dropna().empty:  # Ensure mode() is not empty
        CustomerData[col].fillna(CustomerData[col].mode()[0], inplace=True)
```

```python
CustomerData.drop(columns=['Email', 'Phone', 'Address', 'Time', 'Customer_Segment', 'Zipcode'], inplace=True)
```

```python
CustomerData.isnull().sum()
```

```python
CustomerDataCleaned = CustomerData.dropna()
CustomerDataCleaned.isnull().sum()
```

```python
CustomerDataCleaned.to_csv('RetailCustomerTransactionsCleaned.csv', index=False)
```
